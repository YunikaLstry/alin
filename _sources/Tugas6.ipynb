# EIGEN VALUE & EIGEN VECTOR

**Definisi** 

Jika A adalah $nxn$ maka vektor tidak nol $x$ dinamakan vektor eigen dari A nika $Ax$ adalah kelipan skalar dari X yakni   
  
Ax = $ \lambda X$  
  
Skalar $\lambda$ dinamakan nilai eigen dari A dan x dikatakan vektor eigen yang bersesuaian dengan $\lambda$

Adapun syarat bahwa eigen vektor adalah taknol diberlakukan untuk menghindari kasus yang tidak penting seperti A0 = $\lambda 0$, yang mana berlaku untuk setiap A dan $\lambda$

Ax = λx  
Ax: vektor x setelah transformasi linier  
λx: perkalian bilangan riel dari x, i.e. peregangan dari vektor x.
Transformasi linier untuk eigenvector menghasilkan peregangan vektor
sebanyak eigenvalue kali.

Menghitung eignvalue dan eignvector dengan numpy

```python
# Menghitung eigenvalue dan eigenvector dengan numpy


# matrixs ukuran 2x2
import numpy as np

A = np.array([[3,5], [4,2]])
eign = np.linalg.eig(A)

print("eignvalue: ", eign[0])
print("eignevector: ", eign[1].round())

# matriks ukuran 3 x 3

B = np.array([[2,3,4], [2,4,3], [4,3,2]])
eign2 = np.linalg.eig(B)

print("eignvalue: ", eign2[0])
print("eignvector: ", eign2[1].round())

print()
```

Implementasi eignvalue dan eignvector

```python
import numpy as np


def pcaSVD(matriks):
    row, col = matriks.shape

    matriks_ = matriks - matriks.mean(axis=0)
    u, sigma, vt = np.linalg.svd(matriks_, full_matrices=False)

    
    return ((matriks_ @ vt.T), sigma**2 / (row - 1))

matriks = np.array([[3,5], [4,2]])

eignVector, eignValue = pcaSVD(matriks)
print(eignValue)
print(eignVector.round())
```

## Mencari nilai eigen value dan eigen vector pada matriks 2x2 menggunakan numpy

```python
import numpy as np
from numpy.linalg import eig

# Matriks 2x2
A = np.array([[1,4],[2,3]])

# Menghitung eigenvalue dan eigenvector
w,v = eig(A)

print(A, '\n')
print('E-value', w, '\n')
print('E-value', v.round(), '\n')
```

## Mencari nilai eigen value dan eigen vector pada matriks 3x3 menggunakan numpy

```python
import numpy as np
from numpy.linalg import eig

# Matriks 3x3
A = np.array([[1,4,1],[2,3,2],[2,4,5]])

# Menghitung eigenvalue dan eigenvector
w,v = eig(A)

print(A, '\n')
print('E-value', w.round(), '\n')
print('E-value', v.round(), '\n')
```

## Contoh Implementasi Eigenvalue dan Eigenvector

```python
import numpy as np

def pcaEVD(X):
    n, p = X.shape
    X_centered = X - X.mean(axis=0)
    C = (X_centered.T @ X_centered) / (n - 1)
    e_vals, e_vecs = np.linalg.eig(C)

    idx = e_vals.argsort()[::-1]
    e_vals = e_vals[idx]
    e_vecs = e_vecs[:, idx]

    return (X_centered @ e_vecs), e_vals, e_vecs

X = np.array([[3, 2],
              [2, 3]])

components, eigenvalues, eigenvectors = pcaEVD(X)

rounded_components = np.round(components)
rounded_eigenvalues = np.round(eigenvalues)
rounded_eigenvectors = np.round(eigenvectors)

print("Components:\n", rounded_components)
print("Eigenvalues:\n", rounded_eigenvalues)
print("Eigenvectors:\n", rounded_eigenvectors)
```
